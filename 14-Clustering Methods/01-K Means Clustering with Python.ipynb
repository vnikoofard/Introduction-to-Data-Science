# Introdução ao agrupamento K-Means

K Means Clustering é um algoritmo de aprendizado não supervisionado que tenta agrupar dados com base em sua similaridade. O aprendizado não supervisionado significa que não há resultado a ser previsto e o algoritmo apenas tenta encontrar padrões nos dados. Em *k means clustering*, temos que especificar o número de clusters nos quais queremos que os dados sejam agrupados. O algoritmo atribui aleatoriamente cada observação a um cluster e encontra o centróide de cada cluster. Em seguida, o algoritmo itera através de duas etapas:
Reatribua pontos de dados ao cluster cujo centróide está mais próximo. Calcule o novo centróide de cada cluster. Essas duas etapas são repetidas até que a variação dentro do cluster não possa mais ser reduzida. A variação dentro do cluster é calculada como a soma da distância euclidiana entre os pontos de dados e seus respectivos centróides de cluster.


## Importar as bibliotecas

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Criar um conjuto de dados sintético

```python
from sklearn.datasets import make_blobs
```

```python
# Create Data
data = make_blobs(n_samples=200, n_features=2, 
                           centers=4, cluster_std=1.8, random_state=101)
```

```python
data
```

## Visualizar os dados

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

## Criar os clusters

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

Voce tem que notar, as cores não tem significado entre os dois plots.
