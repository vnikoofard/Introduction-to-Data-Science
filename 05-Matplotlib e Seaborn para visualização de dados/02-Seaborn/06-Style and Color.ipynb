# Estilo e Cor

Mostramos algumas vezes como controlar a estética da figura em seaborn, mas agora vamos analisá-lo formalmente:

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
```

## Estilos

Você pode definir estilos específicos:

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('white')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips,palette='deep')
```

## Remover grades do fundo

```python
sns.countplot(x='sex',data=tips)
sns.despine()
```

```python
sns.countplot(x='sex',data=tips)
sns.despine(left=True)
```

## Tamanho e Aspecto (Size e Aspect)

Você pode usar `plt.figure(figsize=(width,height)` do matplotlib para alterar o tamanho da maioria das plots do Seaborn.

Você pode controlar o tamanho e a proporção da maioria dos gráficos de grade do Seaborn passando os parâmetros: `size` e `aspect`. Por exemplo:

```python

plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python

sns.lmplot(x='total_bill',y='tip',size=2,aspect=4,data=tips)
```

## Escala e contexto

O `set_context()` permite que você substitua os parâmetros padrão:

```python
sns.set_context('poster',font_scale=4)
sns.countplot(x='sex',data=tips,palette='coolwarm')
```
