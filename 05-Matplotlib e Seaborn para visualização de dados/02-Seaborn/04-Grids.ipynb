# Grades (Grids)

Grades são tipos gerais de gráficos que permitem mapear tipos de gráficos para linhas e colunas de uma grade, o que ajuda a criar gráficos semelhantes separados por features.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

## `PairGrid` (Grade de pares)

`Pairgrid` é uma grade de subplot para plotar relacionamentos de pares em um conjunto de dados.

```python
# Just the Grid
sns.PairGrid(iris)
```

```python
# Depois a gente tem que mapear os graficos07
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
# Mapear para upper,lower, and diagonal
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

## `pairplot` (gráfico de pares)

`pairplot` é uma versão mais simples do `PairGrid` (a gente usará com bastante frequência)

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species',palette='rainbow')
```

## `FacetGrid`

`FacetGrid` é a maneira geral de criar grades de gráficos com base em um feature:

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Just the Grid
g = sns.FacetGrid(tips, col="time", row="smoker")
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill")
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker",hue='sex')
# Observe como os argumentos vêm após plt.scatter
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
```

## `JointGrid`

`JointGrid` é a versão geral para grids do tipo `jointplot()`, para um exemplo rápido:

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
```

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot(sns.regplot, sns.distplot)
```

Consulte a documentação conforme necessário para os tipos de grade, mas na maioria das vezes a gente usará apenas os gráficos mais fáceis discutidos anteriormente.
