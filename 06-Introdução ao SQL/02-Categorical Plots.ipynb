# Gráficos de dados categóricos

Agora vamos discutir o uso do seaborn para plotar dados categóricos! Existem alguns tipos de plotagem principais para isso:

* factorplot
* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

Vamos ver exemplos de cada um!

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## `barplot` e `countplot`

Esses gráficos muito semelhantes permitem que você obtenha dados agregados de um "feature" (uma coluna, um dado) categórico em seus dados. `barplot` é um gráfico geral que permite agregar os dados categóricos com base em alguma função, por padrão, a é função média:

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
```

Você pode alterar o objeto estimador para sua própria função, que converte um vetor em um escalar:

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

### `countplot`

Isso é essencialmente o mesmo que barplot, exceto que o estimador está contando explicitamente o número de ocorrências. É por isso que passamos apenas o valor x:

```python
sns.countplot(x='sex',data=tips)
```

## `boxplot` e `violinplot`

boxplots e violinplots são usados para mostrar a distribuição dos dados categóricos. Um gráfico de caixa (ou gráfico "box-and-whisker") mostra a distribuição de dados quantitativos de uma forma que facilita comparações entre variáveis ou entre níveis de uma variável categórica. A caixa (box) mostra os quartis do conjunto de dados enquanto os "bigodes" se estendem para mostrar o resto da distribuição, exceto os pontos que são determinados como “outliers” usando um método que é uma função do intervalo interquartil.

```python
sns.boxplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
# Fazer com o dataset enteiro com orient='h'
sns.boxplot(data=tips,palette='rainbow',orient='h')
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm")
```

### `violinplot`
Um `violinplot` desempenha um papel semelhante ao de um `boxplot`. Ele mostra a distribuição de dados quantitativos em vários níveis de uma (ou mais) variáveis categóricas de forma que essas distribuições possam ser comparadas. Ao contrário de um `boxplot`, no qual todos os componentes do gráfico correspondem a pontos de dados reais, o gráfico de violino apresenta uma estimativa de densidade do kernel da distribuição dos dados.

```python
sns.violinplot(x="day", y="total_bill", data=tips,palette='rainbow')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',palette='Set1')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',split=True,palette='Set1')
```

## `stripplot` e `swarmplot`
O `stripplot` desenhará um gráfico de dispersão (scatter plot) onde uma variável é categórica. Um `stripplot` pode ser desenhado sozinho, mas também é um bom complemento para um `boxplot` ou `violinplot` nos casos em que você deseja mostrar todas as observações junto com alguma representação da distribuição dos dados.

O `swarmplot` é semelhante ao `stripplot`, mas os pontos são ajustados (apenas ao longo do eixo categórico) para que não se sobreponham. Isso dá uma melhor representação da distribuição de valores, embora não seja tão dimensionado para um grande número de observações (tanto em termos da capacidade de mostrar todos os pontos quanto em termos de computação necessária para organizá-los).

```python
sns.stripplot(x="day", y="total_bill", data=tips)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1')
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1',split=True)
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.swarmplot(x="day", y="total_bill",hue='sex',data=tips, palette="Set1", split=True)
```

### Combinando os graficos categoricos

```python
sns.violinplot(x="tip", y="day", data=tips,palette='rainbow')
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=3)
```

## `factorplot`

O `factorplot` é a forma mais geral de um gráfico categórico. Pode receber um parâmetro **kind** para ajustar o tipo de gráfico:

```python
sns.factorplot(x='sex',y='total_bill',data=tips,kind='bar')
```
