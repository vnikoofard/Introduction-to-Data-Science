# Plots de Regressão

O Seaborn tem muitos recursos integrados para gráficos de regressão, no entanto, não discutiremos a regressão até a seção de aprendizado de máquina do curso, portanto, abordaremos apenas a função `lmplot()` por enquanto.

O `lmplot` permite exibir modelos lineares, mas também permite dividir esses gráficos com base nos features, além de colorir o matiz com base nos features.

Vamos explorar como isso funciona:

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## `lmplot()`

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

### Trabalhando com marcadores

`lmplot` kwargs são passados para `regplot` que é uma forma mais geral de lmplot(). regplot tem um parâmetro `scatter_kws` que é passado para `plt.scatter`. Portanto, você deseja definir o parâmetro `s` nesse dicionário, que corresponde (um pouco confuso) ao tamanho do marcador quadrado. Em outras palavras, você acaba passando um dicionário com os argumentos base matplotlib, neste caso, `s` para o tamanho de um scatter plot. Em geral, você provavelmente não se lembrará disso de cabeça, mas, em vez disso, consulte a documentação.

```python
# https://matplotlib.org/stable/api/markers_api.html
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',
           markers=['o','v'],scatter_kws={'s':100})
```

## Usando uma grade (grid)

Podemos adicionar mais separação de variáveis por meio de colunas e linhas com o uso de uma grade. Basta indicar isso com os argumentos `col` ou `row`:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm')
```

## Aspecto e tamanho (`aspect` e `size`)

As figuras Seaborn podem ter seu tamanho e proporção ajustados com os parâmetros `size` e `aspect`:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm',
          aspect=0.6,size=8)
```
