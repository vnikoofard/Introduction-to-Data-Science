## Análise de Vendas

```python
import pandas as pd

dados_vendas = pd.read_csv('https://raw.githubusercontent.com/Uchoa-Gui/DataScience-Projects/main/data/all_data.csv')
dados_vendas
```

```python
colunas_novas = {'Order ID': 'ID', 'Product': 'Produto', 'Quantity Ordered': 'Quantidade', 'Price Each': 'Preco',
                 'Order Date': 'Data', 'Purchase Address': 'Endereco'}
dados_vendas = dados_vendas.rename(colunas_novas, axis=1)
dados_vendas = dados_vendas.drop('Unnamed: 0', axis=1)
```

```python
dados_vendas.info()
```

```python
index = dados_vendas[dados_vendas['Quantidade'] == 'Quantity Ordered'].index
dados_vendas = dados_vendas.drop(index)
```

```python
dados_vendas = dados_vendas.dropna()
```

```python
dados_vendas['Quantidade'] = dados_vendas['Quantidade'].astype(int)
dados_vendas['Preco'] = dados_vendas['Preco'].astype(float)
dados_vendas
```

```python
dados_vendas['Data'] = dados_vendas['Data'].astype('datetime64')
dados_vendas
```

```python
dados_vendas['Mes'] = dados_vendas['Data'].dt.month
dados_vendas
```

```python
import seaborn as sns
agrupado_mes = dados_vendas.groupby('Mes', as_index=False).sum()
sns.barplot(agrupado_mes['Mes'], agrupado_mes['Quantidade'], palette='Blues');
```

```python
dados_vendas.groupby(['Mes', 'Produto'], as_index=False).sum()
```

### 2- Qual cidade vendeu mais?

```python
import matplotlib.pyplot as plt
dados_vendas['cidade'] = dados_vendas['Endereco'].str.split(',').str.get(1)
agrupado_cidade = dados_vendas.groupby('cidade', as_index=False).sum()

sns.barplot(agrupado_cidade['cidade'], agrupado_cidade['Quantidade'])

plt.xticks(rotation=45)
```

### 3- Que horas devemos passar as propagandas de forma a alcançar o maior público?

```python
dados_vendas['hora'] = dados_vendas['Data'].dt.hour 
agrupado_hora = dados_vendas.groupby('hora', as_index=False).sum()

sns.barplot(agrupado_hora['hora'], agrupado_hora['Quantidade'])
```
