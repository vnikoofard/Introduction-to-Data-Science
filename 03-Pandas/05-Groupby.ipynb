# (Agropar) Groupby

O método groupby permite agrupar linhas de dados e chamar funções de agregação

```python
import pandas as pd
# Criar dataframe
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

**Agora você pode usar o método .groupby() para agrupar linhas com base no nome de uma coluna. Por exemplo, vamos agrupar com base na empresa. Isso criará um objeto DataFrameGroupBy:**

```python
df.groupby('Company')
```

Você pode salvar este objeto como uma nova variável:

```python
by_comp = df.groupby("Company")
```

E então chame métodos agregados do objeto:

```python
by_comp.mean()
```

```python
df.groupby('Company').mean()
```

Mais exemplos de métodos agregados:

```python
by_comp.std()
```

```python
by_comp.min()
```

```python
by_comp.max()
```

```python
by_comp.count()
```

```python
by_comp.describe()
```

```python
by_comp.describe().transpose()
```

```python
by_comp.describe().transpose()['GOOG']
```
