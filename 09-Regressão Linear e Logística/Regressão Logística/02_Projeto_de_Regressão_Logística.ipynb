Esse data set contém as seguintes características:

* 'Tempo Diário no Site': é o tempo que o consumindor passou no site, em minutos.

* 'Linha do Topico do Anuncio': é o cabeçalho do anúncio. 
* 'Cidade': Cidade do consumidor.
* 'Tempo do Click': Tempo que o consumidor clicou no anúncio ou que fechou a janela.
* 'Clicou no Anuncio': 0 ou 1 indicam se clicou no anúncio.
