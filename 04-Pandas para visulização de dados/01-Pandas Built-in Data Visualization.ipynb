# Pandas Built-in Data Visualization

In this lecture we will learn about pandas built-in capabilities for data visualization! It's built-off of matplotlib, but it baked into pandas for easier usage!  

Let's take a look!

Nesta seção, aprenderemos sobre os recursos integrados do Pandas para visualização de dados! É construído no matplotlib e plotly, mas é incorporado aos pandas para facilitar o uso!

Vamos dar uma olhada!

## Imports

```python
import numpy as np
import pandas as pd
%matplotlib inline

# Para escolher o Plotly como backend
# pd.options.plotting.backend = "plotly"
```

## Os dados

Existem alguns arquivos csv de dados criados artificialmente que você pode ler como dataframes:

```python
df1 = pd.read_csv('df1',index_col=0)
df2 = pd.read_csv('df2')
```

## Folhas de estilo

Matplotlib tem [style sheets](https://matplotlib.org/stable/gallery/index.html) que você pode usar para deixar seus gráficos um pouco mais bonitos. Essas style sheets incluem `plot_bmh`,`plot_fivethirtyeight`, `plot_ggplot` e muito mais. Eles basicamente criam um conjunto de regras de estilo que seus gráficos seguem. É recomendado usá-los, eles fazem com que todas as suas plots tenham a mesma aparência e sejam mais profissionais. Você pode até criar o seu próprio se quiser que todos os gráficos da sua empresa tenham a mesma aparência (embora seja um pouco tedioso criar).

Aqui está como usá-los.

**Antes de usar `plt.style.use()` seus gráficos ficam assim:**

```python
df1['A'].hist()
```

Chamando o estilo

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

Agora o plot fica nessa forma:

```python
df1['A'].hist()
```

```python
plt.style.use('bmh')
df1['A'].hist()
```

```python
plt.style.use('dark_background')
df1['A'].hist()
```

```python
plt.style.use('fivethirtyeight')
df1['A'].hist()
```

```python
plt.style.use('ggplot')
```

Vamos ficar com o estilo `ggplot` e realmente mostrar como utilizar os recursos de plotagem integrados do pandas!

# Tipos de Lote

Existem vários tipos de gráficos embutidos no Pandas, a maioria deles são gráficos estatísticos por natureza:

* df.plot.area
* df.plot.barh
* df.plot.density
* df.plot.hist
* df.plot.line
* df.plot.scatter
* df.plot.bar
* df.plot.box
* df.plot.hexbin
* df.plot.kde
* df.plot.pie

Você também pode simplesmente chamar `df.plot(kind='hist')` ou substituir esse tipo de argumento por qualquer um dos termos-chave mostrados na lista acima (por exemplo, 'box','barh', etc.)
___

## Area

```python
df2.plot.area(alpha=0.4)
```

## Barplots

```python
df2.head()
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

## Histograms

```python
df1['A'].plot.hist(bins=50)
```

## Line Plots

```python
df1.plot.line(x=df1.index,y='B',figsize=(12,3),lw=1)
```

## Scatter Plots

```python
df1.plot.scatter(x='A',y='B')
```

Você pode usar `c` para colorir com base em outro valor de coluna
Use `cmap` para indicar o mapa de cores a ser usado.
Para todos os mapas de cores, confira: https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

Ou use `s` para indicar o tamanho com base em outra coluna. O parâmetro `s` precisa ser um array, não apenas o nome de uma coluna:

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*200)
```

## BoxPlots

```python
df2.plot.box() # Can also pass a by= argument for groupby
```

## Hexagonal Bin Plot

Útil para dados bivariados, alternativa ao scatter plot:

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='Oranges')
```

____

## Kernel Density Estimation plot (KDE)

```python
df2['a'].plot.kde()
```

```python
df2.plot.density()
```

É isso! Espero que você possa ver por que esse método de plotagem será muito mais fácil de usar do que o matplotlib completo, ele equilibra a facilidade de uso com o controle sobre a figura. Muitas das chamadas de plotagem também aceitam argumentos adicionais do proprio matplotlib
